```python
import numpy as np

# Define A and b
A = np.array([[-6,  2],
              [0, 3]], dtype=float)

b = np.array([-1, 2], dtype=float)

# Solve Ax = b
x = np.linalg.solve(A, b)

print("x =", x)

# Check if any element is > 1 or < 0
if np.any((x > 1) | (x < 0)):
    print("no intersection")
else:
    print("valid intersection")
```
